```python
import urllib.request
from urllib.request import urlopen
from bs4 import BeautifulSoup
import urllib.request
```

```python
htmldata = urlopen('https://leagueoflegends.fandom.com/wiki/Item_(League_of_Legends)')
soup = BeautifulSoup(htmldata, 'html.parser')
images = soup.find_all('img')
itemList = []
```

```python
for index, item in enumerate(images):
    if 'item.png' in str(item):
        itemName = item['data-src'][62:] # starting index of item name, not a magic number!
        itemName = itemName[:itemName.find('/')]
        itemName = ''.join([i for i in itemName if not i.isdigit() and i != '%' and i != '-'])
        itemName = itemName.replace('_',' ').strip()
        itemName = itemName[:-8].strip() # terminal index of item name, no coconut.jpgs here!
        itemList.append(itemName)
        print(itemName)
        directoryStr = 'itemImages\\"'[:-1]
        urllib.request.urlretrieve(str(item['data-src']), directoryStr + itemName+".png")
```
